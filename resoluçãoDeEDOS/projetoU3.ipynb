A resolução de equações diferenciais ordinárias (EDOs) é um dos tópicos centrais da matemática aplicada, com vastas aplicações em diversas áreas do conhecimento, como física, engenharia, entre outras. Embora a resolução exata de muitas EDOs seja desejada, na prática, muitas dessas equações não têm soluções analíticas, ou seja, não é possível encontrar uma expressão fechada para a solução. Nestes casos, é necessário recorrer a métodos numéricos que proporcionem aproximações da solução.

Dois dos métodos numéricos mais populares e utilizados para a resolução de EDOs são o Método de Euler e o Método de Runge-Kutta. Ambos são métodos de integração numérica que permitem calcular aproximações sucessivas das soluções de EDOs, mas apresentam características, vantagens e limitações distintas.

Para a execução deste trabalho, foi-se proposto inicialmente o uso do Método de Euler para a resolução das EDOs; porém, durante um período de estudo e análise dos métodos, encontrou-se também o Método de Range-Kutta, que obteve resultados satisfatórios quanto ao erro.

#### Descrição:
O Método de Euler é um dos métodos mais simples e fundamentais para resolver EDOs. A ideia principal desse método é aproximar a solução da equação diferencial usando uma aproximação linear em pequenos passos de tempo.

#### Características:
- **Simplicidade**: O Método de Euler é extremamente simples de implementar e entender.
- **Precisão**: A precisão do método depende do tamanho do passo \( h \). O erro global do método é da ordem \( O(h) \), o que significa que a precisão melhora linearmente com a redução do tamanho do passo.
- **Estabilidade**: O Método de Euler pode ser instável, especialmente para EDOs rígidas, onde o erro pode crescer rapidamente em grandes intervalos de tempo.

#### Vantagens:
- Fácil de implementar.
- Computacionalmente barato.

#### Desvantagens:
- A precisão pode ser baixa, especialmente para grandes intervalos de tempo.
- O método pode ser instável, a menos que o passo seja suficientemente pequeno.

### 2. Método de Runge-Kutta

#### Descrição:
O Método de Runge-Kutta (RK) é uma família de métodos mais avançados para a solução de EDOs. O mais comum e utilizado é o **Método de Runge-Kutta de quarta ordem (RK4)**. O método RK4 tenta melhorar a precisão do Método de Euler calculando uma média ponderada de várias estimativas da derivada em cada passo.

#### Características:
- **Precisão**: O Método de Runge-Kutta de quarta ordem tem um erro global da ordem \( O(h^4) \), o que significa que a precisão melhora de forma muito mais significativa do que o Método de Euler ao reduzir o passo \( h \).
- **Complexidade**: O método RK4 é mais complexo que o Euler, pois envolve calcular várias estimativas de derivadas por passo de integração, o que aumenta o custo computacional.
- **Estabilidade**: O método RK4 é mais estável que o Método de Euler e pode ser usado com passos maiores sem perder tanto em precisão.

#### Vantagens:
- Alta precisão (erro de ordem \( O(h^4) \)).
- Estabilidade em uma ampla gama de problemas.
- Mais robusto para EDOs rígidas em comparação com o Método de Euler.

#### Desvantagens:
- Mais complexo de implementar.
- Custo computacional maior, pois envolve mais cálculos por passo.

---
Como podemos ver aqui, o método de Range-Kutta, embora mais custoso para se processar, ofereceria resultados mais precisose e é mais robusto em relação a EDOs rígidas. É amplamente utilizado em situações onde precisão é crucial e o tempo de computação não é uma limitação.

Abaixo, construí um gráfico em notação de Big O para esboçar o erro de ambos os métodos:


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir os valores de h (tamanho do passo)
h_values = np.logspace(-1, -5, 100)  # Valores de h entre 10^-1 e 10^-5

# Calcular o erro global para o Método de Euler (O(h)) e Runge-Kutta (O(h^4))
error_euler = h_values  # erro global do Método de Euler é proporcional a h
error_rk4 = h_values**4  # erro global do RK4 é proporcional a h^4

# Criar o gráfico
plt.figure(figsize=(8, 6))

# Plotar os erros
plt.loglog(h_values, error_euler, label="Método de Euler (O(h))", color="blue", linestyle='--')
plt.loglog(h_values, error_rk4, label="Método de Runge-Kutta 4 (O(h^4))", color="red", linestyle='-')

# Adicionar rótulos e título
plt.xlabel("Tamanho do passo (h)", fontsize=12)
plt.ylabel("Erro Global", fontsize=12)
plt.title("Comparação de Erros: Método de Euler vs Método de Runge-Kutta", fontsize=14)

# Adicionar legenda
plt.legend()

# Mostrar o gráfico
plt.grid(True)
plt.show()
```

O gráfico gerado com o código compara o erro global de dois métodos numéricos: o Método de Euler e o Método de Runge-Kutta de quarta ordem (RK4), em relação ao tamanho do passo ℎ.

### Comparação: Método de Euler vs Runge-Kutta
Abaixo, montei uma tabela rápida com as informações obtidas durante minhas pesquisas para resolução deste trabalho

| Característica             | Método de Euler                         | Método de Runge-Kutta (RK4)                |
|----------------------------|-----------------------------------------|--------------------------------------------|
| **Simplicidade**            | Muito simples de implementar.           | Mais complexo devido ao cálculo de múltiplos \( k \)'s. |
| **Precisão**                | Baixa (erro de ordem \( O(h) \)).       | Alta (erro de ordem \( O(h^4) \)).         |
| **Custo Computacional**     | Baixo (menos cálculos por passo).       | Maior (mais cálculos por passo).           |
| **Estabilidade**            | Pode ser instável para grandes \( h \). | Mais estável, especialmente para EDOs rígidas. |
| **Uso Comum**              | Problemas simples ou com passos muito pequenos. | Problemas mais complexos ou quando maior precisão é necessária. |

### Conclusão

- **Método de Euler**: É um método simples e eficiente em termos computacionais, mas sua precisão é limitada. O erro global é da ordem \( O(h) \), o que significa que a solução numérica pode ser significativamente imprecisa se o passo \( h \) não for pequeno o suficiente.
  
- **Método de Runge-Kutta de quarta ordem**: Embora mais complexo e custoso computacionalmente, o Runge-Kutta é muito mais preciso, com um erro global da ordem \( O(h^4) \). Isso significa que, mesmo com passos maiores, o Runge-Kutta pode fornecer soluções muito mais precisas do que o Método de Euler, o que o torna a escolha preferencial para muitos problemas de EDOs.

A escolha entre esses métodos depende do equilíbrio entre **precisão** e **custo computacional**. Se a precisão for crucial, o Runge-Kutta será preferido, enquanto que para problemas mais simples ou quando o tempo de computação for uma restrição, o Método de Euler pode ser suficiente.

```python
import numpy as np
import matplotlib.pyplot as plt

def sistema_dinamico(t, y, m1, m2, k1, k2, b1, b2, F):
    x1, v1, x2, v2 = y
    dx1dt = v1
    dv1dt = (-k1*x1 - k2*(x1 - x2) - b1*v1 - b2*(v1 - v2)) / m1
    dx2dt = v2
    dv2dt = (-k2*(x2 - x1) - b2*(v2 - v1) + F) / m2
    return np.array([dx1dt, dv1dt, dx2dt, dv2dt])

# Parâmetros do sistema
m1, m2 = 20, 20
k1, k2 = 1000, 2000
b1, b2 = 1, 5
F = 10

# Condições iniciais
y = np.array([0.1, 0, -0.1, 0])

# Parâmetros de tempo
t_final = 20
dt = 0.000001
n_steps = int(t_final/dt)
t = np.linspace(0, t_final, n_steps)

# Arrays para armazenar resultados
results = np.zeros((4, n_steps))
results[:,0] = y

# Método de Euler
for i in range(1, n_steps):
    dy = sistema_dinamico(t[i-1], results[:,i-1], m1, m2, k1, k2, b1, b2, F)
    results[:,i] = results[:,i-1] + dt * dy

x1, v1, x2, v2 = results

# Plots
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, (1, 2))
plt.plot(t, x1, 'b-', label='Massa 1 (x₁)', linewidth=2)
plt.plot(t, x2, 'r-', label='Massa 2 (x₂)', linewidth=2)
plt.grid(True, alpha=0.3)
plt.xlabel('Tempo (s)')
plt.ylabel('Deslocamento (m)')
plt.title('Resposta do Sistema - Deslocamentos')
plt.legend()

# Criar grade para as setinhas
x1_grid = np.linspace(min(x1), max(x1), 20)
v1_grid = np.linspace(min(v1), max(v1), 20)
X1, V1 = np.meshgrid(x1_grid, v1_grid)

# Calcular as direções das setinhas (neste caso, usando um modelo simples)
U1 = V1
V1_arrows = -X1  # Modelo simplificado de oscilador

# Normalizar os vetores
norm1 = np.sqrt(U1**2 + V1_arrows**2)
U1 = U1/norm1
V1_arrows = V1_arrows/norm1

plt.subplot(2, 2, 3)
plt.quiver(X1, V1, U1, V1_arrows, alpha=0.3, color='gray')
plt.plot(x1, v1, 'b-', label='Massa 1', linewidth=2)
plt.grid(True, alpha=0.3)
plt.xlabel('Posição (m)')
plt.ylabel('Velocidade (m/s)')
plt.title('Diagrama de Fases - Massa 1')
plt.axhline(y=0, color='k', linestyle='--', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='--', alpha=0.3)
plt.legend()

# Criar grade para as setinhas
x2_grid = np.linspace(min(x2), max(x2), 20)
v2_grid = np.linspace(min(v2), max(v2), 20)
X2, V2 = np.meshgrid(x2_grid, v2_grid)

# Calcular as direções das setinhas
U2 = V2
V2_arrows = -X2  # Modelo simplificado de oscilador

# Normalizar os vetores
norm2 = np.sqrt(U2**2 + V2_arrows**2)
U2 = U2/norm2
V2_arrows = V2_arrows/norm2

plt.subplot(2, 2, 4)
plt.quiver(X2, V2, U2, V2_arrows, alpha=0.3, color='gray')
plt.plot(x2, v2, 'r-', label='Massa 2', linewidth=2)
plt.grid(True, alpha=0.3)
plt.xlabel('Posição (m)')
plt.ylabel('Velocidade (m/s)')
plt.title('Diagrama de Fases - Massa 2')
plt.axhline(y=0, color='k', linestyle='--', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='--', alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Como o resto dos parâmetros, condições e variáveis para modelar o sistema já foram definidas previamente,
# Vamos partir do ponto onde o código se modifica

from scipy.integrate import solve_ivp

# Tempo de simulação
t_span = (0, 20)  # 10 segundos de simulação
t_eval = np.linspace(*t_span, 1000)

# Resolver o sistema
sol = solve_ivp(sistema_dinamico, t_span, y, t_eval=t_eval,args=(m1, m2, k1, k2, b1, b2, F),method='RK45')
# Método de Runge-Kutta de 4ª/5ª ordem, definido em method='RK45'

# Extrair resultados
t = sol.t
x1, v1, x2, v2 = sol.y

# Plots
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, (1, 2))
plt.plot(t, x1, 'b-', label='Massa 1 (x₁)', linewidth=2)
plt.plot(t, x2, 'r-', label='Massa 2 (x₂)', linewidth=2)
plt.grid(True, alpha=0.3)
plt.xlabel('Tempo (s)')
plt.ylabel('Deslocamento (m)')
plt.title('Resposta do Sistema - Deslocamentos')
plt.legend()

# Criar grade para as setinhas
x1_grid = np.linspace(min(x1), max(x1), 20)
v1_grid = np.linspace(min(v1), max(v1), 20)
X1, V1 = np.meshgrid(x1_grid, v1_grid)

# Calcular as direções das setinhas (neste caso, usando um modelo simples)
U1 = V1
V1_arrows = -X1  # Modelo simplificado de oscilador

# Normalizar os vetores
norm1 = np.sqrt(U1**2 + V1_arrows**2)
U1 = U1/norm1
V1_arrows = V1_arrows/norm1

plt.subplot(2, 2, 3)
plt.quiver(X1, V1, U1, V1_arrows, alpha=0.3, color='gray')
plt.plot(x1, v1, 'b-', label='Massa 1', linewidth=2)
plt.grid(True, alpha=0.3)
plt.xlabel('Posição (m)')
plt.ylabel('Velocidade (m/s)')
plt.title('Diagrama de Fases - Massa 1')
plt.axhline(y=0, color='k', linestyle='--', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='--', alpha=0.3)
plt.legend()

# Criar grade para as setinhas
x2_grid = np.linspace(min(x2), max(x2), 20)
v2_grid = np.linspace(min(v2), max(v2), 20)
X2, V2 = np.meshgrid(x2_grid, v2_grid)

# Calcular as direções das setinhas
U2 = V2
V2_arrows = -X2  # Modelo simplificado de oscilador

# Normalizar os vetores
norm2 = np.sqrt(U2**2 + V2_arrows**2)
U2 = U2/norm2
V2_arrows = V2_arrows/norm2

plt.subplot(2, 2, 4)
plt.quiver(X2, V2, U2, V2_arrows, alpha=0.3, color='gray')
plt.plot(x2, v2, 'r-', label='Massa 2', linewidth=2)
plt.grid(True, alpha=0.3)
plt.xlabel('Posição (m)')
plt.ylabel('Velocidade (m/s)')
plt.title('Diagrama de Fases - Massa 2')
plt.axhline(y=0, color='k', linestyle='--', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='--', alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```

De cara, percebe-se que visualmente os resultados se mostram bem similares aos obtidos pelo Método de Euler.

Agora que temos os resultados do método de Runge-Kutta de quarta/quinta ordem (method='RK45') utilizando o solve_ivp da biblioteca SciPy, podemos comparar os resultados obtidos com o Método de Euler e o Método de Runge-Kutta.

Comparação dos Resultados

À medida que o tempo passa, a solução do Método de Euler pode começar a mostrar distorções mais evidentes.

Em comparação, o Range-Kutta tem um erro muito menor devido à sua ordem de precisão mais alta, sendo mais eficaz em simulações de longo prazo.
```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do sistema
a, b = 1, 0.25
c, d = 0.01, 0.02
e, h = 0.02, 0

def sistema(t, z):
    x, y = z
    dxdt = x * (a - c * x - d * y)
    dydt = -y * (b - e * x) - h
    return np.array([dxdt, dydt])

# Parâmetros de simulação
t_final = 100
dt = 0.00001
n_steps = int(t_final/dt)
t = np.linspace(0, t_final, n_steps)

# Arrays para resultados
z = np.zeros((2, n_steps))
z[:,0] = [10, 5]  # Condições iniciais

# Método de Euler
for i in range(1, n_steps):
    dz = sistema(t[i-1], z[:,i-1])
    z[:,i] = z[:,i-1] + dt * dz

# Gráficos
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(t, z[0], label='População de presas (x)')
plt.xlabel('Tempo')
plt.ylabel('População de presas')
plt.title('População de Presas vs Tempo')
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(t, z[1], label='População de predadores (y)', color='orange')
plt.xlabel('Tempo')
plt.ylabel('População de predadores')
plt.title('População de Predadores vs Tempo')
plt.grid()

# Criar grade para as setinhas
x_grid = np.linspace(min(z[0]), max(z[0]), 20)
y_grid = np.linspace(min(z[1]), max(z[1]), 20)
X, Y = np.meshgrid(x_grid, y_grid)

U = a*X - b*X*Y  # variação da população de presas
V = -c*Y + d*X*Y  # variação da população de predadores

# Normalizar os vetores
norm = np.sqrt(U**2 + V**2)
U = U/norm
V = V/norm

plt.subplot(2, 2, 3)
plt.quiver(X, Y, U, V, alpha=0.3, color='gray')
plt.plot(z[0], z[1], label='Diagrama de Fase', color='green')
plt.xlabel('População de presas')
plt.ylabel('População de predadores')
plt.title('Diagrama de Fase')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
L, L0 = 2.0, 1.0
k, m = 40.0, 1.0
g = 9.81
beta = 0.2

# Condições iniciais
theta0 = 3 * np.pi / 4
v0 = 0.0

def pendulum_amortecido(t, y):
    x1, x2 = y
    dx1dt = x2
    dx2dt = -(g / L) * np.sin(x1) - beta * x2
    return np.array([dx1dt, dx2dt])

# Parâmetros temporais
t_final = 25
dt = 0.00001
n_steps = int(t_final/dt)
t = np.linspace(0, t_final, n_steps)

# Arrays para resultados
y = np.zeros((2, n_steps))
y[:,0] = [theta0, v0]

# Método de Euler
for i in range(1, n_steps):
    dy = pendulum_amortecido(t[i-1], y[:,i-1])
    y[:,i] = y[:,i-1] + dt * dy

# Gráficos
plt.figure(figsize=(14, 8))

plt.subplot(2, 1, 1)
plt.plot(t, y[0], label='Posição (θ)', color='blue', linewidth=2)
plt.plot(t, y[1], label='Velocidade (ω)', color='red', linestyle='--', linewidth=2)
plt.title('Posição e Velocidade em função do Tempo')
plt.xlabel('Tempo (s)')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)

# Criar grade para as setinhas
theta_grid = np.linspace(min(y[0]), max(y[0]), 30)
omega_grid = np.linspace(min(y[1]), max(y[1]), 30)
THETA, OMEGA = np.meshgrid(theta_grid, omega_grid)
U = OMEGA  # variação de θ (velocidade angular)
V = -(g/L)*np.sin(THETA) - b*OMEGA  # variação de ω (aceleração angular)

# Normalizar os vetores
norm = np.sqrt(U**2 + V**2)
U = U/norm
V = V/norm

plt.subplot(2, 1, 2)
plt.quiver(THETA, OMEGA, U, V, alpha=0.3, color='gray')
plt.plot(y[0], y[1], label='Diagrama de Fase', color='green', linewidth=2)
plt.title('Diagrama de Fase')
plt.xlabel('Posição (θ)')
plt.ylabel('Velocidade (ω)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

Ao ajustar o parâmetro de amortecimento 𝛽 = 0.2, garantimos que o pêndulo permanece em um regime de subamortecimento. Diferente do regime de sobreamortecimento, o pêndulo não volta imediatamente para a posição de equilíbrio, mas leva algum tempo para que as oscilações se extingam, refletindo a característica do subamortecimento de uma dissipação gradual da energia.

```python
from scipy.integrate import solve_ivp

# Condições iniciais
y0 = [theta0, v0]

# Intervalo de tempo para a simulação
t_span = (0, 20)  # Simulação de 20 segundos
t_eval = np.linspace(0, 20, 1000)  # Mais pontos para suavizar a curva

# Solução numérica
sol = solve_ivp(pendulum_amortecido, t_span, y0, t_eval=t_eval)

# Gráficos
plt.figure(figsize=(14, 8))

# Gráfico da posição (ângulo) e velocidade em função do tempo
plt.subplot(2, 1, 1)
plt.plot(sol.t, sol.y[0], label='Posição (θ) - Ângulo (rad)', color='blue', linewidth=2)
plt.plot(sol.t, sol.y[1], label='Velocidade (ω) - Velocidade angular (rad/s)', color='red', linestyle='--', linewidth=2)
plt.title('Posição e Velocidade em função do Tempo', fontsize=14)
plt.xlabel('Tempo (s)', fontsize=12)
plt.ylabel('Valor', fontsize=12)
plt.legend(loc='upper right')
plt.grid(True)

# Diagrama de fase (Posição vs. Velocidade)
# Criar grade para as setinhas
theta_grid = np.linspace(min(y[0]), max(y[0]), 30)
omega_grid = np.linspace(min(y[1]), max(y[1]), 30)
THETA, OMEGA = np.meshgrid(theta_grid, omega_grid)
U = OMEGA  # variação de θ (velocidade angular)
V = -(g/L)*np.sin(THETA) - b*OMEGA  # variação de ω (aceleração angular)

# Normalizar os vetores
norm = np.sqrt(U**2 + V**2)
U = U/norm
V = V/norm

plt.subplot(2, 1, 2)
plt.quiver(THETA, OMEGA, U, V, alpha=0.3, color='gray')
plt.plot(y[0], y[1], label='Diagrama de Fase', color='green', linewidth=2)
plt.title('Diagrama de Fase')
plt.xlabel('Posição (θ)')
plt.ylabel('Velocidade (ω)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
# Mostramos aqui os resultados obtidos
```

Ambas as abordagens, tanto o Euler quanto o Runge-Kutta, mostram a tendência do pêndulo a se aproximar do ponto de equilíbrio (𝜃=0, 𝜔 = 0), mas o método de Runge-Kutta oferece uma aproximação mais confiável desse comportamento ao longo do tempo.

Assim como os resultados vistos nas outras questões, para esses casos de simulações rápidas e simples, ambos obtiveram uma precisão excelente para leitura e interpretação de dados, mostrando que a escolha pelo Método de Euler para modelagens para "primeiras análise" é mais que o suficiente.
